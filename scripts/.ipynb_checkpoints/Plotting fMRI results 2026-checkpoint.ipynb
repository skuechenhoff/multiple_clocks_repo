```python
# helper functions

def pair_correct_tasks(data_dict, keys_list):
    """
    data_dict: dict with keys like 'A1_forw_A_reward'
    keys_list: ordered list of keys you want to include and in what order
    Returns two matrices: one for the first element of each pair, one for its match.
    """
    # Define task pairing relationships
    task_pairs = {'1_forw': '2_backw', '1_backw': '2_forw'}
    th_1, th_2, paired_list_control  = [], [], []
    # Loop through keys in the *specified order*
    for key in keys_list:
        assert key in data_dict, "Missmatch between model rdm keys and data RDM keys"
        task, direction, state, phase = key.split('_')  # e.g. ['A1', 'forw', 'A', 'reward']
        # Create the pairing suffix (e.g. from '1_forw' → '2_backw')
        pair_suffix = task_pairs.get(f"{task[-1]}_{direction}")
        # Build the paired key (e.g. 'A2_backw_A_reward')
        pair_key = f"{task[0]}{pair_suffix}_{state}_{phase}"
        # Only add if both keys exist
        if pair_key in data_dict:
            th_1.append(np.asarray(data_dict[key]))
            th_2.append(np.asarray(data_dict[pair_key]))
            paired_list_control.append(f"{key} with {pair_key}")

    # import pdb; pdb.set_trace()       
    th_1 = np.vstack(th_1)
    th_2 = np.vstack(th_2)
    # print(paired_list_control)
    return th_1, th_2, paired_list_control



def vec_to_mat(vec):
    L = len(vec)
    n = int((np.sqrt(8 * L + 1) - 1) / 2)
    M = np.full((n, n), np.nan, dtype=float)
    iu = np.triu_indices(n)
    M[iu] = vec
    return M




import numpy as np
import re

def n_from_triu_len(L: int, k: int = 0) -> int:
    """
    Infer n from length L of np.triu_indices(n, k=k).
    For k=0: L = n(n+1)/2
    """
    # Works for k=0 (your case). If you ever change k, handle separately.
    n = int((np.sqrt(8 * L + 1) - 1) / 2)
    if n * (n + 1) // 2 != L:
        raise ValueError(f"Length {L} is not n(n+1)/2 for any integer n (got n={n}).")
    return n

def unflatten_triu(vec, n=None, k: int = 0, symmetric: bool = True):
    """
    Reconstruct an (n,n) matrix from flattened upper-triangle (including diag for k=0).
    """
    vec = np.asarray(vec)
    if n is None:
        n = n_from_triu_len(len(vec), k=k)
    M = np.zeros((n, n), dtype=float)
    iu = np.triu_indices(n, k=k)
    M[iu] = vec
    if symmetric:
        # mirror upper to lower; keep diagonal unchanged
        M = M + M.T - np.diag(np.diag(M))
    return M

def parse_paired_label(lbl: str):
    """
    Example:
      'A1_backw_A_path with A2_forw_A_path'
    -> block = 'A1 ⟵  |  A2 ⟶'
       within = 'A-path' (or 'A-rew', etc.)
    """
    left, right = lbl.split(" with ")
    l = left.split("_")
    r = right.split("_")
    # l: [A1, backw, A, path] ; r: [A2, forw, A, path]
    arrow = {"backw": "⟵", "forw": "⟶"}
    block = f"{l[0]} {arrow.get(l[1], l[1])}  |  {r[0]} {arrow.get(r[1], r[1])}"

    within = f"{l[2]}-{l[3]}"
    within = within.replace("reward", "rew")  # shorter
    return block, within


import matplotlib.pyplot as plt
import matplotlib as mpl

def plot_rdm(
    rdm_flat,
    paired_labels,
    title="RDM",
    block_size=8,
    show="hier",          # "hier" or "full"
    cmap="RdBu_r",        # good “power” diverging palette for correlations
    center=0.0,           # center colormap at 0 for similarity/correlation
    vmin=None,
    vmax=None,
    draw_diag=False,
    figsize=(10, 9),
):
    """
    show="hier":
      - axis ticks: only 1 label per block (every 8), placed at block centers
      - adds a small key explaining within-block order (A-path, A-rew, ...)
    show="full":
      - uses all 80 paired_labels on x/y (rotated). Can get busy.
    """
    M = unflatten_triu(rdm_flat, symmetric=True)

    n = M.shape[0]
    if len(paired_labels) != n:
        raise ValueError(f"len(paired_labels)={len(paired_labels)} but matrix is {n}x{n}")

    if not draw_diag:
        np.fill_diagonal(M, np.nan)

    # Parse hierarchical labels
    parsed = [parse_paired_label(l) for l in paired_labels]
    block_labels = [parsed[i][0] for i in range(0, n, block_size)]
    within_labels = [parsed[i][1] for i in range(0, block_size)]

    # Norm: diverging centered at `center` if data crosses it
    data_min = np.nanmin(M) if vmin is None else vmin
    data_max = np.nanmax(M) if vmax is None else vmax
    if vmin is None: vmin = data_min
    if vmax is None: vmax = data_max

    if (vmin < center) and (vmax > center):
        norm = mpl.colors.TwoSlopeNorm(vmin=vmin, vcenter=center, vmax=vmax)
    else:
        norm = mpl.colors.Normalize(vmin=vmin, vmax=vmax)

    fig, ax = plt.subplots(figsize=figsize, constrained_layout=True)
    im = ax.imshow(M, cmap=cmap, norm=norm, interpolation="nearest", aspect="equal")

    # White separators every block_size
    for b in range(block_size, n, block_size):
        ax.axhline(b - 0.5, color="white", lw=1.6, alpha=0.95)
        ax.axvline(b - 0.5, color="white", lw=1.6, alpha=0.95)

    # Ticks + labels
    if show == "full":
        ax.set_xticks(np.arange(n))
        ax.set_yticks(np.arange(n))
        ax.set_xticklabels(paired_labels, rotation=90, fontsize=6)
        ax.set_yticklabels(paired_labels, fontsize=6)
    elif show == "hier":
        centers = np.arange(block_size/2 - 0.5, n, block_size)
        ax.set_xticks(centers)
        ax.set_yticks(centers)
        ax.set_xticklabels(block_labels, rotation=90, fontsize=9)
        ax.set_yticklabels(block_labels, fontsize=9)

        # Within-block key (order repeats for every block)
        key = "Within-block order (repeats in every 8):  " + "  |  ".join(within_labels)
        ax.set_xlabel(key, fontsize=9, labelpad=12)
    else:
        raise ValueError("show must be 'hier' or 'full'")

    ax.set_title(title, fontsize=12, pad=10)
    ax.tick_params(length=0)

    cbar = fig.colorbar(im, ax=ax, fraction=0.046, pad=0.02)
    cbar.set_label("Similarity", rotation=90)

    return fig, ax
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

def n_from_triu_len(L: int, k: int = 0) -> int:
    # supports k=0 (incl diag) and k=1 (excl diag)
    if k == 0:
        n = int((np.sqrt(8 * L + 1) - 1) / 2)
        if n * (n + 1) // 2 != L:
            raise ValueError(f"Bad length {L} for k=0.")
        return n
    if k == 1:
        n = int((1 + np.sqrt(1 + 8 * L)) / 2)
        if n * (n - 1) // 2 != L:
            raise ValueError(f"Bad length {L} for k=1.")
        return n
    raise NotImplementedError("Only k=0 or k=1 supported here.")

def unflatten_triu_upper_only(vec, n=None, k: int = 0, draw_diag: bool = True):
    """
    Returns an (n,n) matrix with only the stored upper-triangle filled.
    Everything else is NaN (will plot white).
    """
    vec = np.asarray(vec)
    if n is None:
        n = n_from_triu_len(len(vec), k=k)

    M = np.full((n, n), np.nan, dtype=float)
    iu = np.triu_indices(n, k=k)
    M[iu] = vec

    if not draw_diag:
        np.fill_diagonal(M, np.nan)
    return M

def parse_paired_label(lbl: str):
    left, right = lbl.split(" with ")
    l = left.split("_")   # [A1, backw, A, path]
    r = right.split("_")  # [A2, forw, A, path]
    arrow = {"backw": "←", "forw": "→"}
    block = f"{l[0]}{arrow.get(l[1], '')} | {r[0]}{arrow.get(r[1], '')}"
    within = f"{l[2]}-{l[3]}".replace("reward", "rew")
    return block, within

def plot_rdm_upper(
    rdm_flat,
    paired_labels,
    title="RDM",
    block_size=8,
    k=0,
    draw_diag=False,
    cmap="RdBu_r",
    center=0.0,
    vmin=None,
    vmax=None,
    # small-figure label controls
    label_mode="blocks",      # "blocks" | "full" | "none"
    label_on="x",             # "x" | "y" | "both"
    label_fontsize=6,
    figsize=(4.2, 4.2),
    show_within_key=False,    # keep off for tiny figures; turn on for big ones
):
    M = unflatten_triu_upper_only(rdm_flat, n=len(paired_labels), k=k, draw_diag=draw_diag)

    # Make NaNs render as white
    cmap_obj = plt.get_cmap(cmap).copy()
    cmap_obj.set_bad(color="white")

    # Norm (centered if it spans center)
    finite = np.isfinite(M)
    if vmin is None: vmin = np.nanmin(M[finite]) if finite.any() else -1
    if vmax is None: vmax = np.nanmax(M[finite]) if finite.any() else  1
    if (vmin < center) and (vmax > center):
        norm = mpl.colors.TwoSlopeNorm(vmin=vmin, vcenter=center, vmax=vmax)
    else:
        norm = mpl.colors.Normalize(vmin=vmin, vmax=vmax)

    fig, ax = plt.subplots(figsize=figsize, constrained_layout=True)
    im = ax.imshow(np.ma.masked_invalid(M), cmap=cmap_obj, norm=norm, interpolation="nearest", aspect="equal")
    ax.set_facecolor("white")

    n = M.shape[0]

    # block separators (still useful even though lower triangle is white)
    for b in range(block_size, n, block_size):
        ax.axhline(b - 0.5, color="white", lw=1.2, alpha=0.95)
        ax.axvline(b - 0.5, color="white", lw=1.2, alpha=0.95)

    # labels
    parsed = [parse_paired_label(l) for l in paired_labels]
    block_labels = [parsed[i][0] for i in range(0, n, block_size)]

    def set_axis_labels(axis):
        if label_mode == "none":
            if axis == "x":
                ax.set_xticks([])
            else:
                ax.set_yticks([])
            return

        if label_mode == "full":
            ticks = np.arange(n)
            labels = paired_labels
        else:  # "blocks"
            ticks = np.arange(block_size/2 - 0.5, n, block_size)
            labels = block_labels

        if axis == "x":
            ax.set_xticks(ticks)
            ax.set_xticklabels(labels, rotation=90, fontsize=label_fontsize)
        else:
            ax.set_yticks(ticks)
            ax.set_yticklabels(labels, fontsize=label_fontsize)

    if label_on in ("x", "both"):
        set_axis_labels("x")
    else:
        ax.set_xticks([])

    if label_on in ("y", "both"):
        set_axis_labels("y")
    else:
        ax.set_yticks([])

    ax.tick_params(length=0, pad=1)
    ax.set_title(title, fontsize=9)

    # optional within-block key (only nice on larger figs)
    if show_within_key and block_size <= n:
        within = [parsed[i][1] for i in range(0, block_size)]
        ax.set_xlabel("Within-block order: " + " | ".join(within), fontsize=label_fontsize, labelpad=6)

    # small colorbar (optional; can also turn off for grids)
    cbar = fig.colorbar(im, ax=ax, fraction=0.045, pad=0.02)
    cbar.ax.tick_params(labelsize=label_fontsize)
    return fig, ax
```
